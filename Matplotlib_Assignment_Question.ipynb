```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
students = ["Aman", "Riya", "Rahul", "Sneha"]
marks = [80, 90, 75, 85]
plt.plot(students , marks , marker='o', color='green', linestyle='-')

plt.title('Students mark analysis' )
plt.xlabel('Students')
plt.ylabel('marks')
```

```python
brands = ["Samsung", "Apple", "MI", "OnePlus", "Vivo"]
sales = [50, 40, 70, 30, 60]
plt.bar(brands , sales , color=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"])
```

```python
activities = ["Study", "Sleep", "Games", "Exercise"]
hours = [6, 8, 5, 2]
plt.pie(hours , labels=activities, autopct='%1.1f%%', startangle=140) 
```

```python
hours = [1, 2, 3, 4, 5, 6]
marks = [30, 40, 50, 60, 70, 85]
plt.scatter(hours , marks , color='green',)
```

```python
days = [1, 2, 3, 4, 5]
aman = [60, 65, 70, 75, 80]
riya = [55, 60, 68, 72, 78]

plt.plot(days, aman, label="Aman", linestyle="-", marker="o", color="blue")
plt.plot(days, riya, label="Riya", linestyle="--", marker="s", color="orange")

plt.xlabel("Days")
plt.ylabel("Performance Marks")

plt.legend()
```

```python
marks = [45, 56, 67, 78, 89, 90, 56, 67, 76, 88, 92]
plt.hist(marks , bins = 5 , edgecolor = 'black' , color="purple")

plt.title("Distribution of Marks")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 25, 18, 32, 22]
plt.plot(x , y , color="teal" , marker="D" , markersize=8, label="Growth Trend")
plt.title("Customized Growth Performance Chart", fontsize=14, fontweight="bold")
plt.xlabel("Timeline (Months)", fontsize=12)
plt.ylabel("Value (in Thousands)", fontsize=12)

plt.grid(True, linestyle="solid", alpha=0.5, color="gray")

plt.legend()
```

```python
players = ["Virat", "Rohit", "Dhoni", "Gill"]
runs = [82, 75, 50, 90]
plt.bar(players , runs, color=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"])
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [30, 32, 35, 33, 31, 29, 28]

plt.plot(days, temperature, marker="o", linestyle="-", color="crimson", linewidth=2)

plt.title("Weekly Weather Analysis (Temperature Trend)", fontweight = 'bold')
plt.xlabel("Days of the Week")
plt.ylabel("Temperature (°C)")

plt.grid(True, linestyle=":", alpha=0.6)
```
