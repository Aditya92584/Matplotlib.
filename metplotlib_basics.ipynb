```python
# plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'], batsman['V Kohli'],color = 'green',marker = 'o',markersize = '4',label = 'virat')
plt.plot(batsman['index'], batsman['RG Sharma'],color = 'red',marker = 'o',markersize = '4',label = 'Rohit')
plt.title('Rohit Vs Virat Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc ='upper left')
plt.grid()
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2016'],color = ('#01084f','#391954','#631e50','#a73c5a','#ff7954'))
```

```python
x = np.arange(df.shape[0])

plt.bar(x - 0.2, df['2015'], width=0.2, color='red',edgecolor = 'black')
plt.bar(x, df['2016'], width=0.2, color='yellow',edgecolor = 'black')
plt.bar(x + 0.2, df['2017'], width=0.2, color='green',edgecolor = 'black')

plt.show()
```

```python
plt.bar(df['batsman'], df['2015'], label = '2015',edgecolor = 'black')
plt.bar(df['batsman'], df['2016'], bottom = df['2015'], label = '2016',edgecolor = 'black')
plt.bar(df['batsman'], df['2017'], bottom = (df['2015'] + df['2016']), label = '2017',edgecolor = 'black')

plt.legend()
plt.show()
```
